```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
%matplotlib inline
```

```python
time_feature = pd.read_csv("ouput/20DRY-DS")
time_feature.head(3)
```

```python
update = lambda x: x[0:10]
time_feature['time'] = time_feature['datetime'].map(update)
```

```python
time_feature.head(3)
```

```python
x = time_feature.groupby('time').sum()
```

```python
formatter = '%Y-%m-%d'

def datetime_complete(df):
  "complement the missing time units with zero value"
  global formatter
  data = []
  previousDatetime = None
  c = None
  for number, currentDatetime in df.itertuples(index = False):
    c = datetime.datetime.strptime(currentDatetime, formatter)
    if (previousDatetime != None):
      p = datetime.datetime.strptime(previousDatetime, formatter)
      complementZero(data, p, c)
    previousDatetime = currentDatetime
    data.append([currentDatetime, number])
  res = pd.DataFrame(data)
  res.columns = df.columns.values
  return res

def complementZero(data, start, end):
  global formatter
  delta = int((end - start).total_seconds() // 86400)
  if (delta > 1):
    for i in range(1, delta):
      start = start + datetime.timedelta(days = 1)
      data.append([start.strftime(formatter), 0])
```

```python
c = datetime_complete(x)
c.head(3)
```

```python
month_7 = c[0:22]
month_8 = c[22:53]
month_9 = c[53:83]
month_10 = c[83:114]
month_11 = c[114:144]
month_12 = c[144:175]
month_1 = c[175:206]
month_2 = c[206:234]
month_3 = c[234:265]
```
